```python
# Final version
import os
import glob
import pandas as pd
from tqdm import tqdm

# Set the directory containing the CSV files
csv_dir = './merge_folder_csv_active_incident_resource_summary'

# Create the output directory if it doesn't exist
output_dir = './merge_folder_csv_active_incident_resource_summary'
if not os.path.exists(output_dir):
    os.makedirs(output_dir)

# Get a list of all CSV files in the directory
csv_files = glob.glob(os.path.join(csv_dir, '*.csv'))

# Combine all CSV files into a single dataframe
combined_df = pd.concat([pd.read_csv(f) for f in tqdm(csv_files)])

# Save the combined dataframe to a new CSV file in the output directory
output_file = os.path.join(output_dir, 'combined_file2.csv')
combined_df.to_csv(output_file, index=False)
```
